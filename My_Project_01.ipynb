```python
# import the libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("2022_forbes_billionaires.csv")
data["networth"]=data["networth"].str.strip("$")
data["networth"]=data["networth"].str.strip("B")
data["networth"]=data["networth"].astype(float)
data.head(12)
```

df=data.sort_values(by=["networth"],ascending=False).head(10)

```python
df=data.sort_values(by=["networth"],ascending=False).head(10)
plt.figure(figsize=(20,10))
plt.bar(df['name'],df['networth']) 
plt.show()
```

# Top 5 domains with the most number of billionaires


```python
a=data["source"].value_counts().head(5)
index=a.index
print(index)
sources=a.values
print(sources)
col=["cyan","violet","yellow","skyblue","pink"]
plt.figure(figsize=(5,5))
plt.pie(sources,labels=index,colors=col)
center=plt.Circle((0,0),0.3,color="white")
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("Top 5 domains with the most number of billionaires",fontsize=20)
plt.show()
```

# Top 5 industries with the most number of billionaires

```python
a=data["industry"].value_counts().head(5)
index=a.index
print(index)
industry=a.values
print(industry)
col=["grey","purple","brown","orange","magenta"]
plt.figure(figsize=(5,5))
plt.pie(industry,labels=index,colors=col)
center=plt.Circle((0,0),0.2,color="white")
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("Top 5 industry with the most number of billionaires",fontsize=20)
plt.show()
```

# Top 5 countries with the most number of billionaires

```python
a=data["country"].value_counts().head(5)
index=a.index
print(index)
country=a.values
print(country)
col=["gold","pink","silver","blue","black"]
plt.figure(figsize=(5,5))
plt.pie(country,labels=index,colors=col)
center=plt.Circle((0,0),0.3,color="white")
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("Top 5 industry with the most number of billionaires",fontsize=20)
plt.show()
```
